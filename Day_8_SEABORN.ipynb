```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sns.set_theme(style='whitegrid')
```

```python
# 1 Relational Plot 
tips = sns.load_dataset("tips")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", style="smoker", size = "size")
plt.title("scatter plot of total bill vs tip")
plt.show()
```

```python
fmri = sns.load_dataset("fmri")
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")
plt.title("FMRI Signal Over Time")
plt.show()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="day", kind="scatter")
plt.show()
```

```python
# 2 Distribution Plots
sns.histplot(data=tips, x="total_bill", kde = True, bins=20, hue="sex", multiple="stack")
plt.title("distribution of total bill amounts")
plt.show()
```

```python
sns.kdeplot(data=tips, x ="total_bill", hue = "time", fill=True, alpha=0.3)
plt.title("density estimate of total bill")
plt.show()
```

```python
sns.ecdfplot(data=tips, x="total_bill", hue="time")
plt.title("ECDF of total bill")
plt.show()
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
plt.show()
```

```python
sns.jointplot(data=tips, x="total_bill", y="tip", kind="reg", hue="smoker")
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue="species")
plt.show()
```

```python
# 3 categorial plots

sns.catplot(data=tips, x="day", y="total_bill", kind="box", hue="sex")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="total_bill", hue="smoker")
plt.title("Box Plot of Total Bill by Day")
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.title("Violin Plot of Total Bill by Day")
plt.show()
```

```python
sns.boxenplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title("Boxen Plot of Total Bill by Day")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex", ci="sd")
plt.title("Average Total Bill by Day")
plt.show()
```

```python
sns.countplot(data=tips, x="day", hue="sex")
plt.title("Count of Customers by Day")
plt.show()
```

```python
sns.pointplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Point Plot of Total Bill by Day")
plt.show()
```

```python
sns.stripplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Strip Plot of Total Bill by Day")
plt.show()
```

```python
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.title("Swarm Plot of Total Bill by Day")
plt.show()
```

```python
# Regression Plots
sns.regplot(data=tips, x="total_bill", y="tip", ci=95, scatter_kws={"s": 80})
plt.title("Regression Plot of Tip vs Total Bill")
plt.show()
```

```python
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", col="time")
plt.show()
```

```python
sns.residplot(data=tips, x="total_bill", y="tip")
plt.title("Residual Plot")
plt.show()
```

```python
# 5 matrix plots
flights = sns.load_dataset("flights")
flights_pivot = flights.pivot("month", "year", "passengers")
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passengers Heatmap")
plt.show()
```

```python
sns.clustermap(flights_pivot, standard_scale=1, cmap="YlGnBu")
plt.show()
```

```python
# 6 Multi-plot Grids
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

```python
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot_joint(sns.scatterplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
# 7. Time Series Plots
flights = sns.load_dataset("flights")
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
plt.title("May Passengers Over Years")
plt.show()
```

```python
# 8. Theme and Aesthetics
# Available styles: darkgrid, whitegrid, dark, white, ticks
sns.set_style("whitegrid")

# Context settings: paper, notebook, talk, poster
sns.set_context("notebook")

# Color palettes
sns.set_palette("husl")

# Reset defaults
sns.reset_defaults()
```

```python
# Display a color palette
sns.palplot(sns.color_palette("husl", 8))

# Create and display a diverging palette
sns.palplot(sns.diverging_palette(220, 20, n=7))

# Create and display a cubehelix palette
sns.palplot(sns.cubehelix_palette(8))
```

```python
# 9. Advanced Plot Customization
g = sns.scatterplot(data=tips, x="total_bill", y="tip")
g.set_title("Total Bill vs Tip")
g.set_xlabel("Total Bill ($)")
g.set_ylabel("Tip ($)")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
g = sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.savefig("scatterplot.png", dpi=300, bbox_inches="tight")
```

```python
# 10. Statistical Estimation
from numpy import median

sns.barplot(data=tips, x="day", y="total_bill", hue="sex", estimator=median)
plt.title("Median Total Bill by Day")
plt.show()
```

```python
sns.barplot(data=tips, x="day", y="total_bill", ci=68)  # 68% confidence interval
plt.show()
```

```python
# 11. Plotting Wide-Form Data
# Create wide-form data
index = pd.date_range("1/1/2020", periods=100)
wide_df = pd.DataFrame(np.random.randn(100, 5), index=index, columns=list("ABCDE"))

# Plot all columns
sns.lineplot(data=wide_df)
plt.title("Multiple Time Series")
plt.show()
```

```python
# 12. Combining with Matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(data=tips, x="total_bill", y="tip", ax=ax1)
ax1.set_title("Scatter Plot")

sns.boxplot(data=tips, x="day", y="total_bill", ax=ax2)
ax2.set_title("Box Plot")

plt.tight_layout()
plt.show()
```

```python
# 13. Specialized Plots
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill")
plt.title("KDE with Rug Plot")
plt.show()
```

```python
# 14. Working with Long-Form vs Wide-Form Data
sns.boxplot(data=tips, x="day", y="total_bill")
```

```python
wide_data = tips.pivot(columns="day", values="total_bill")
sns.boxplot(data=wide_data)
```

```python
# 15. Plotting Large Datasets
diamonds = sns.load_dataset("diamonds")
sns.kdeplot(data=diamonds, x="carat", y="price", thresh=0.1, cmap="viridis")
plt.title("Density of Carat vs Price")
plt.show()
```

```python
sns.histplot(data=diamonds, x="carat", y="price", bins=50, cbar=True)
plt.title("Binned Carat vs Price")
plt.show()
```
